# Data Visualization

```python
import seaborn as sns
```

```python
sns.pairplot(data=iris_df,hue='Species',diag_kind='kde')
```

```python
x=iris_df.iloc[:,[0,1,2,3]].values
x
```

```python
 iris_results=pd.crosstab(index=iris["Species"],columns="count")
iris_results
```

```python
type_i=iris_df.Species.values
type_i[0:10]
```

```python
#target variable
from sklearn.preprocessing import LabelEncoder 
k=LabelEncoder()
y=k.fit_transform(type_i)
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
# Defining the decision tree algorithm
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,Y_train)

print('Decision Tree Classifer Created')
```

```python
predict=dtree.predict(X_test)
predict
```

```python
#visualizing  Decision Tree 
import matplotlib.pyplot as plt
import numpy as np
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(dtree,feature_names=np.array(iris_df.columns),filled=True)
```

```python
from sklearn import metrics
```

```python
dt=pd.DataFrame({"Actual": Y_test,"Predicted":predict})
dt
```

```python
error=metrics.accuracy_score(Y_test,predict)
error
```
